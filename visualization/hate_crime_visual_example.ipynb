```python
import pandas as pd
import altair as alt
```

```python
unemployment_bronx = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERBronxCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_kings = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERKingsCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_newyork = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERNewYorkCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_queens = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERQueensCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_richmond = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERRichmondCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
```

```python
unemployments = [unemployment_bronx,
                 unemployment_kings,
                 unemployment_newyork,
                 unemployment_queens,
                 unemployment_richmond]
```

```python
for i, unemployment in enumerate(unemployments):
    unemployment['Date'] = pd.to_datetime(unemployment['Date'], format='%Y-%m')
    unemployments[i] = unemployment.set_index('Date')
```

```python
unemployment_nyc = sum(unemployments) / 5
```

```python
unemployment_nyc = unemployment_nyc[:22].reset_index()
```

```python
unemployment_nyc
```

```python
unemployment_chart = alt.Chart(unemployment_nyc).mark_line(color='red').encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ]
)
```

# Hate Crimes Output

```python
hate_crimes = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/hate_crimes_daily_out.csv")
```

```python
hate_crimes
```

```python
hate_crimes['record_date'] = pd.to_datetime(hate_crimes['record_date'], format="%m/%d/%Y")
```

```python
hate_crimes = hate_crimes.set_index('record_date')
```

```python
hate_crimes = hate_crimes.sort_index()
```

```python
hate_crimes = hate_crimes.resample('M').sum()
```

```python
hate_crimes = hate_crimes.reset_index()
```

```python
hate_crimes
```

```python
hate_crime_chart = alt.Chart(hate_crimes).transform_fold(
    fold=['anti_white', 'anti_asian', 'anti_black', 'anti_jewish', 'anti_hispanic'],
    as_=['races', 'num']
).mark_bar(size=20).encode(
    x='record_date:T',
    y='num:Q',
    color='races:N',
    strokeDash='races:N',
    tooltip=[
        'record_date:T',
        'num:Q',
        'races:N',
    ],
    
)
```

```python
alt.layer(
    hate_crime_chart,
    unemployment_chart,
).properties(
    width=700,
    height=500,
).interactive()
```

```python
import pandas as pd
import altair as alt
```

```python
unemployment_bronx = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERBronxCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
```

```python
import re
```

```python
unemployment_bronx = unemployment_bronx[~unemployment_bronx['Date'].str.endswith('00')]
```

```python
unemployment_bronx['Date'] = pd.to_datetime(unemployment_bronx['Date'], format='%Y-%m')
```

```python
unemployment_bronx.dtypes
```

```python
alt.Chart(unemployment_bronx).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ]
).interactive()
```

# Hate Crimes Output

```python
hate_crimes = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/hate_crimes_daily_out.csv")
```

```python
hate_crimes
```

```python
hate_crimes['record_date'] = pd.to_datetime(hate_crimes['record_date'], format="%m/%d/%Y")
```

```python
hate_crimes = hate_crimes.set_index('record_date')
```

```python
hate_crimes = hate_crimes.sort_index()
```

```python
hate_crimes = hate_crimes.resample('M').sum()
```

```python
hate_crimes = hate_crimes.reset_index()
```

```python
hate_crimes
```

```python
alt.Chart(hate_crimes).transform_fold(
    fold=['anti_white', 'anti_asian', 'anti_black', 'anti_jewish', 'anti_hispanic'],
    as_=['races', 'num']
).mark_line().encode(
    x='record_date:T',
    y='num:Q',
    color='races:N',
    strokeDash='races:N',
).properties(
    width=700,
    height=400
).configure_legend(
    labelFontSize=24,
    titleFontSize=24,
).interactive()
```

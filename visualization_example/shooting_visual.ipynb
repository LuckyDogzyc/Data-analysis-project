```python
import pandas as pd
import altair as alt
```

```python
unemployment_bronx = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERBronxCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_kings = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERKingsCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_newyork = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERNewYorkCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_queens = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERQueensCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
unemployment_richmond = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/UERRichmondCounty.out"
                                 , header=None, names=['Date', 'Umemployment Rate'])
```

```python
unemployment_bronx.dtypes
unemployment_kings.dtypes
unemployment_newyork.dtypes
unemployment_queens.dtypes
unemployment_richmond.dtypes
```

```python
unemployments = [unemployment_bronx,
                 unemployment_kings,
                 unemployment_newyork,
                 unemployment_queens,
                 unemployment_richmond]
```

```python
for i, unemployment in enumerate(unemployments):
    unemployment['Date'] = pd.to_datetime(unemployment['Date'], format='%Y-%m')
    unemployments[i] = unemployment.set_index('Date')
```

```python
unemployment_bronx['Date'] = pd.to_datetime(unemployment_bronx['Date'], format='%Y-%m')
unemployment_kings['Date'] = pd.to_datetime(unemployment_kings['Date'], format='%Y-%m')
unemployment_newyork['Date'] = pd.to_datetime(unemployment_newyork['Date'], format='%Y-%m')
unemployment_queens['Date'] = pd.to_datetime(unemployment_queens['Date'], format='%Y-%m')
unemployment_richmond['Date'] = pd.to_datetime(unemployment_richmond['Date'], format='%Y-%m')
unemployment_total['Date'] = pd.to_datetime(unemployment_total['Date'], format='%Y-%m')
```

```python
plot_bronx = alt.Chart(unemployment_bronx).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ],
    color=alt.value('blue')
).interactive()

plot_kings = alt.Chart(unemployment_kings).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ],
    color=alt.value('green')
).interactive()

plot_newyork = alt.Chart(unemployment_newyork).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ],
    color=alt.value('orange')
).interactive()

plot_queens = alt.Chart(unemployment_queens).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ],
    color=alt.value('yellow')
).interactive()

plot_richmond = alt.Chart(unemployment_richmond).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ],
    color=alt.value('brown')
).interactive()

plot_total = alt.Chart(unemployment_total).mark_line().encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ],
    color=alt.value('red')
).interactive()

plot_bronx + plot_kings + plot_newyork + plot_queens + plot_richmond
```

```python
#2019 Unemployment Rate
unemployment_nyc = sum(unemployments) / 5
```

```python
unemployment_2019 = unemployment_nyc[10:22].reset_index()
```

```python
unemployment_chart_19 = alt.Chart(unemployment_2019).mark_line(color='red').encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ]
)
unemployment_chart_19
```

```python
#2020 Unemployment Rate
unemployment_2020 = unemployment_nyc[:10].reset_index()
```

```python
unemployment_chart_20 = alt.Chart(unemployment_2020).mark_line(color='red').encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ]
)
unemployment_chart_20
```

```python
#2019-2020 Unemployment Rate
unemployment_2019_2020 = unemployment_nyc[:22].reset_index()
unemployment_chart_19_20 = alt.Chart(unemployment_2019_2020).mark_line(color='red').encode(
    x='Date',
    y='Umemployment Rate',
    tooltip=[
        'Date:T',
        'Umemployment Rate:Q'
    ]
)
unemployment_chart_19_20
```

# Shooting Output

```python
shootingData2020 = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/ShootingCount2020.out"
                               , header=None, names=['Date', 'Count'])
shootingDataPast = pd.read_csv("https://raw.githubusercontent.com/ZhenZhouXu/Data-analysis-project/main/data_sources/outputs/ShootingCountPast.out"
                               , header=None, names=['Date', 'Count'])
```

```python
shootingData2020 = shootingData2020[:-1]
```

```python
shootingDataPast = shootingDataPast[-309:-1]
```

```python
#2019-2020 Shooting Cases
line_past = alt.Chart(shootingDataPast).mark_line(
    color='blue',
    size=1
).transform_window(
    rolling_mean='mean(Count)',
    frame=[-15, 15]
).encode(
    x='Date:T',
    y=alt.Y('Count:Q',
            axis=alt.Axis(title='Shooting Cases'))
)
line_2020 = alt.Chart(shootingData2020).mark_line(
    color='blue',
    size=1
).transform_window(
    rolling_mean='mean(Count)',
    frame=[-15, 15]
).encode(
    x='Date:T',
    y=alt.Y('Count:Q',
            axis=alt.Axis(title='Shooting Cases'))
)
line = line_past + line_2020
```

```python
#2019-2020 Shooting Cases and Unemployment Rate
line + unemployment_chart_19_20
```
